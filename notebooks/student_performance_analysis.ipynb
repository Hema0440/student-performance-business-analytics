# Student Performance Business Analytics

This notebook analyzes student performance data and provides
business insights and actionable recommendations.


```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/student_marks.csv")
df.head()
```

```python
# Create Pass/Fail column based on business rule
df["Result"] = df.apply(
    lambda row: "Pass" if row["Maths"] >= 40 and row["Science"] >= 40 and row["English"] >= 40 else "Fail",
    axis=1
)

df[["Name", "Maths", "Science", "English", "Result"]]
```

```python
subject_avg = df[["Maths", "Science", "English"]].mean()
subject_avg
```

```python
subject_avg.plot(kind="bar", title="Average Marks by Subject")
```

## Business Insights – Subject Performance

- The analysis shows clear variation in average marks across subjects.
- The subject with the lowest average score indicates a need for additional academic support and focused intervention.
- The subject with the highest average score reflects effective teaching methods that can be replicated across other subjects.


```python
gender_avg = df.groupby("Gender")[["Maths", "Science", "English"]].mean()
gender_avg
```

```python
gender_avg.plot(kind="bar", title="Gender-wise Average Marks")
```

## Business Insights – Gender Performance

- The analysis shows differences in academic performance between male and female students.
- One gender group demonstrates slightly higher average scores across most subjects.
- These insights can help institutions design targeted mentoring and support programs.


## Business Recommendations

- Allocate additional academic resources and faculty support to subjects with lower average performance.
- Introduce remedial classes for students identified as failing to improve overall pass rates.
- Implement targeted mentoring programs based on gender-wise performance insights.
- Regularly monitor student performance using data analysis to support continuous improvement.
